# Penguin Analysis

```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df =sns.load_dataset('penguins')
df.head(5)
```

```python
df = df.dropna(how='any')
df.info()
```

```python
x =df['bill_depth_mm'].values
x
```

## BODY MASS ACROSS SPECIES

```python
#ax = sns.boxplot(x = 'species', y = 'body_mass_g', data = dataframe)

# oof, that's ugly lets reset size
sns.set(rc={'figure.figsize': (14, 5)})
sns.set_style("ticks")
sns.set_context("talk")

# might need a style update 
#sns.set_style('ticks')

ax = sns.boxplot(x = 'species', y = 'body_mass_g', data = df, palette = ['gold', 'aqua', 'pink'])
ax = sns.swarmplot(x = 'species', y = 'body_mass_g', data = df, color = 'red')
#print(df['body_mass_g'].mean())
```

## BODY MASS ACROSS ISLANDS

```python
sns.set(rc={'figure.figsize': (14, 5)})
sns.set_context("poster", font_scale = 0.5, rc={"grid.linewidth": 1.0})

# might need a style update 
#sns.set_style('ticks')

ax = sns.boxplot(x = 'island', y = 'body_mass_g', data = df)
ax = sns.swarmplot(x = 'island', y = 'body_mass_g', data = df, color = 'red')
```

Biscoe island has penguins with larger body mass as compared to the other two islands. 

The distribution is most even in Torgersen island.

## Splitting plots with column wrapping

## Correlation of flipper length and body mass across islands

```python
cx = sns.lmplot(x = "flipper_length_mm", y = "body_mass_g", data = df, height = 5, 
                hue = 'island', col = 'island', markers = ['+', '*', 'p'], col_wrap = 3, ci = False)
```

Biscoe islandhas a strong positive correlation of flipper and body mass. 

This indicates that as the bigger penguins have large flippers.
